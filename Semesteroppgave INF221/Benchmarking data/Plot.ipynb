```python
import pandas 
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt


plt.rcParams['axes.titlesize'] = 9
plt.rcParams['axes.labelsize'] = 9
plt.rcParams['xtick.labelsize'] = 8
plt.rcParams['ytick.labelsize'] = 8
plt.rcParams['legend.fontsize'] = 8
plt.rcParams['text.usetex'] = True

def new_figure(height=55):
    "Return figure with width 84mm and given height in mm."

    return plt.figure(figsize=(84/25.4, height/25.4))
```

```python
quick_sort = pandas.read_pickle("quick_sort",compression=None)
```

```python
merge_sort = pandas.read_pickle("merge_sort", compression=None)
```

```python
heap_sort = pandas.read_pickle("heap_sort", compression=None)
```

```python
python_sorted = pandas.read_pickle("python_sorted", compression=None)
```

```python
numpy_sort = pandas.read_pickle("numpy_sort", compression=None)
```

```python
data_quick1 = {'Size':[10, 100, 1000, 10000, 100000], 
       'random_list_int':quick_sort[1][2],
       'random_list_float':quick_sort[1][3], 
       'random_charlist':quick_sort[1][4] } 

data_quick2 = {'Size':[10, 100, 1000],'ascending_list_int': quick_sort[1][0], 'descending_list_int': quick_sort[1][1]}

data_merge = {'Size':[10, 100, 1000, 10000, 100000], 'ascending_list_int':merge_sort[1][0],
       'descending_list_int':merge_sort[1][1], 
       'random_list_int':merge_sort[1][2] ,
       'random_list_float':merge_sort[1][3], 
       'random_charlist':merge_sort[1][4]} 

data_heap = {'Size':[10, 100, 1000, 10000, 100000], 'ascending_list_int':heap_sort[1][0],
       'descending_list_int':heap_sort[1][1], 
       'random_list_int':heap_sort[1][2] ,
       'random_list_float':heap_sort[1][3], 
       'random_charlist':heap_sort[1][4]} 

data_python = {'Size':[10, 100, 1000, 10000, 100000, 1000000, 10000000], 'ascending_list_int':python_sorted[1][0],
       'descending_list_int':python_sorted[1][1], 
       'random_list_int':python_sorted[1][2] ,
       'random_list_float':python_sorted[1][3], 
       'random_charlist':python_sorted[1][4]} 

data_numpy = {'Size':[10, 100, 1000, 10000, 100000, 1000000, 10000000], 'ascending_list_int':numpy_sort[1][0],
       'descending_list_int':numpy_sort[1][1], 
       'random_list_int':numpy_sort[1][2] ,
       'random_list_float':numpy_sort[1][3], 
       'random_charlist':numpy_sort[1][4]} 
  
```

```python
df_quick1 = pd.DataFrame(data_quick1) 

df_quick2 = pd.DataFrame(data_quick2) 

df_merge = pd.DataFrame(data_merge)

df_heap = pd.DataFrame(data_heap)

df_python = pd.DataFrame(data_python)

df_numpy = pd.DataFrame(data_numpy)
```

```python
fig = new_figure()
ax = fig.add_subplot(1, 1, 1)


ax = sns.lineplot(x = 'Size', y = 'ascending_list_int', data=df_quick2, label = "Ouicksort")
ax = sns.lineplot(x = 'Size', y = 'ascending_list_int', data=df_merge, label = "Merge sort")
ax = sns.lineplot(x = 'Size', y = 'ascending_list_int', data=df_heap, label = "Heap sort")
ax = sns.lineplot(x = 'Size', y = 'ascending_list_int', data=df_python, label = "Python sorted")
ax = sns.lineplot(x = 'Size', y = 'ascending_list_int', data=df_numpy, label = "Numpy sort")



ax.set_xscale('log')
ax.set_yscale('log')
y = plt.ylabel('Benchmark Time (s)')
x = plt.xlabel('Testdata size')
y = plt.title('Testdata: Ascending List Int')

ax.legend()
fig.savefig('ascending_list_int_plot.pdf', bbox_inches='tight')


```

```python
fig = new_figure()
ax = fig.add_subplot(1, 1, 1)

ax = sns.lineplot(x = 'Size', y = 'descending_list_int', data=df_quick2, label = "Ouicksort")
ax = sns.lineplot(x = 'Size', y = 'descending_list_int', data=df_merge, label = "Merge sort")
ax = sns.lineplot(x = 'Size', y = 'descending_list_int', data=df_heap, label = "Heap sort")
ax = sns.lineplot(x = 'Size', y = 'descending_list_int', data=df_python, label = "Python sorted")
ax = sns.lineplot(x = 'Size', y = 'descending_list_int', data=df_numpy, label = "Numpy sort")

#x = plt.axis(xmin=0, xmax=1000, ymin=0, ymax=0.500)
ax.set_xscale('log')
ax.set_yscale('log')
y = plt.ylabel('Benchmark Time (s)')
x = plt.xlabel('Testdata size')
y = plt.title('Testdata: Descending List Int')

ax.legend()
fig.savefig('descending_list_int_plot.pdf', bbox_inches='tight')
```

```python
fig = new_figure()
ax = fig.add_subplot(1, 1, 1)

ax = sns.lineplot(x = 'Size', y = 'random_list_int', data=df_quick1, label = "Ouicksort")
ax = sns.lineplot(x = 'Size', y = 'random_list_int', data=df_merge, label = "Merge sort")
ax = sns.lineplot(x = 'Size', y = 'random_list_int', data=df_heap, label = "Heap sort")
ax = sns.lineplot(x = 'Size', y = 'random_list_int', data=df_python, label = "Python sorted")
ax = sns.lineplot(x = 'Size', y = 'random_list_int', data=df_numpy, label = "Numpy sort")

ax.set_xscale('log')
ax.set_yscale('log')
y = plt.ylabel('Benchmark Time (s)')
x = plt.xlabel('Testdata size')
y = plt.title('Testdata: Random List Int')

ax.legend()
fig.savefig('random_list_int_plot.pdf', bbox_inches='tight')


```

```python
fig = new_figure()
ax = fig.add_subplot(1, 1, 1)
ax = sns.lineplot(x = 'Size', y = 'random_list_float', data=df_quick1, label = "Ouicksort")
ax = sns.lineplot(x = 'Size', y = 'random_list_float', data=df_merge, label = "Merge sort")
ax = sns.lineplot(x = 'Size', y = 'random_list_float', data=df_heap, label = "Heap sort")
ax = sns.lineplot(x = 'Size', y = 'random_list_float', data=df_python, label = "Python sorted")
ax = sns.lineplot(x = 'Size', y = 'random_list_float', data=df_numpy, label = "Numpy sort")

ax.set_xscale('log')
ax.set_yscale('log')
y = plt.ylabel('Benchmark Time (s)')
x = plt.xlabel('Testdata size')
y = plt.title('Testdata: Random List Float')
ax.legend()
fig.savefig('random_list_float_plot.pdf', bbox_inches='tight')
```

```python
fig = new_figure()
ax = fig.add_subplot(1, 1, 1)
ax = sns.lineplot(x = 'Size', y = 'random_charlist', data=df_quick1, label = "Ouicksort")
ax = sns.lineplot(x = 'Size', y = 'random_charlist', data=df_merge, label = "Merge sort")
ax = sns.lineplot(x = 'Size', y = 'random_charlist', data=df_heap, label = "Heap sort")
ax = sns.lineplot(x = 'Size', y = 'random_charlist', data=df_python, label = "Python sorted")
ax = sns.lineplot(x = 'Size', y = 'random_charlist', data=df_numpy, label = "Numpy sort")

ax.set_xscale('log')
ax.set_yscale('log')
y = plt.ylabel('Benchmark Time (s)')
x = plt.xlabel('Testdata size')
y = plt.title('Testdata: Random List Charlist')
ax.legend()
fig.savefig('random_charlist_plot.pdf', bbox_inches='tight')
```
